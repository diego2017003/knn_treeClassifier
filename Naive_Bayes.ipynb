## Prática com Naive Bayes

Nesse notebook iremos utilizar 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
dados_originais = pd.read_csv('https://raw.githubusercontent.com/diego2017003/knn_treeClassifier/main/heart.csv?token=AHICD55RZU4G53PRUPRVB3DA4HQCS')
dados_amostra30 = pd.read_csv('https://raw.githubusercontent.com/diego2017003/knn_treeClassifier/main/amostra_30_sr.csv?token=AHICD57VCQG2JACG2W6CH6DA4HQEE')
dados_amostra50 = pd.read_csv('https://raw.githubusercontent.com/diego2017003/knn_treeClassifier/main/amostra_50_estratificada.csv?token=AHICD577TFMYIROBCPV67PDA4HQHY')
dados_amostra70 = pd.read_csv('https://raw.githubusercontent.com/diego2017003/knn_treeClassifier/main/amostra_70_estratificada.csv?token=AHICD52TMNHP253IGVDF243A4HQP4')
```

```python
dados_originais_tratados = dados_originais.copy()
dados_originais_tratados.restecg.replace({0:'Normal',1:'Possui_anormalidade',2:'Provável_hipertrofia'}, inplace=True)
dados_originais_tratados.cp.replace({1:'Angina_tipica',2:'Angina_atipica',3:'Não_Angina',0:'Assintomatico'}, inplace=True)
dados_originais_tratados = pd.get_dummies(dados_originais_tratados , columns=['restecg','cp'] , prefix=['restecg_','cp_'])
dados_originais_tratados.describe()
```

```python
dados_amostra30.drop(columns=['Unnamed: 0'],inplace=True)
dados_amostra50.drop(columns=['Unnamed: 0'],inplace=True)
dados_amostra70.drop(columns=['Unnamed: 0'],inplace=True)
dados_amostra30.head()
```

Iremos aplicar o método de naive bayes com kernel Gaussiano devido a característica contínua dos nossos dados.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

### Dados Originais
---

```python
acuracia =[]
Naive_Bayes_gaussiano = GaussianNB()
X = dados_originais_tratados.drop(columns=['output'])
X_train,X_test,Y_train,Y_test = train_test_split(X,dados_originais_tratados.output,test_size=0.3,
                                                 stratify=dados_originais_tratados.output)
Naive_Bayes_gaussiano.fit(X_train,Y_train)
y_pred = Naive_Bayes_gaussiano.predict(X_test)
acuracia1 = accuracy_score(Y_test,y_pred)
print("Acuracia 70/30 :",acuracia1)
acuracia.append(acuracia1)
Naive_Bayes_gaussiano2 = GaussianNB()
X = dados_originais_tratados.drop(columns=['output'])
X_train,X_test,Y_train,Y_test = train_test_split(X,dados_originais_tratados.output,test_size=0.2,
                                                 stratify=dados_originais_tratados.output)
Naive_Bayes_gaussiano.fit(X_train,Y_train)
y_pred = Naive_Bayes_gaussiano.predict(X_test)
acuracia2 = accuracy_score(Y_test,y_pred)
print("Acuracia 80/20 :",acuracia2)
acuracia.append(acuracia2)
Naive_Bayes_gaussiano3 = GaussianNB()
X = dados_originais_tratados.drop(columns=['output'])
X_train,X_test,Y_train,Y_test = train_test_split(X,dados_originais_tratados.output,test_size=0.1,
                                                 stratify=dados_originais_tratados.output)
Naive_Bayes_gaussiano.fit(X_train,Y_train)
y_pred = Naive_Bayes_gaussiano.predict(X_test)
acuracia3 = accuracy_score(Y_test,y_pred)
print("Acuracia 90/10 :",acuracia3)
acuracia.append(acuracia3)
y = dados_originais_tratados['output']
GNB = GaussianNB()
scores = cross_val_score(GNB, X, y, cv=10)
print("kfold, 10 folds :",scores.mean())  
print("acuracia média ", np.mean(acuracia))
print("desvio padrão ", np.std(acuracia))
```
